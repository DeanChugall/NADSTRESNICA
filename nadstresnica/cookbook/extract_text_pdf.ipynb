```python
import os

import pytesseract
from PIL import Image

# Set Tesseract path if necessary (Windows users)
# Uncomment and change the path based on your installation
# pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"


def extract_text(image_path, output_text_file="extracted_text.txt"):
    """
    Extracts text from an image using Tesseract OCR and saves it to a text file.

    :param image_path: Path to the image file.
    :param output_text_file: Path to save extracted text.
    """
    try:
        # Open the image
        image = Image.open(image_path)

        # Perform OCR
        extracted_text = pytesseract.image_to_string(image, lang="srp")  # Use "eng" for English text

        # Save the extracted text
        with open(output_text_file, "w", encoding="utf-8") as f:
            f.write(extracted_text)

        print(f"Text successfully extracted and saved to {output_text_file}")

    except Exception as e:
        print(f"Error: {e}")


# Example usage:
image_path = "your_image.png"  # Change this to the path of your image
extract_text(image_path)
```

```python
import os

import pytesseract
from pdf2image import convert_from_path
from PIL import Image

# Postavljanje TESSDATA_PREFIX na tačnu putanju
os.environ["TESSDATA_PREFIX"] = "/usr/share/tesseract-ocr/4.00/tessdata/"


def izdvoji_tekst_iz_skeniranog_pdf(pdf_fajl, izlazni_txt_fajl="izdvojeni_tekst1.txt"):
    """
    Ekstrahuje tekst iz skeniranog PDF-a pomoću OCR-a i čuva ga u tekstualnom fajlu.

    :param pdf_fajl: Putanja do PDF fajla.
    :param izlazni_txt_fajl: Putanja za čuvanje izdvojenog teksta.
    """
    try:
        # Konvertovanje PDF-a u slike
        stranice = convert_from_path(pdf_fajl)

        # Obrada svake stranice kroz OCR
        izdvojeni_tekst = ""
        for broj_stranice, stranica in enumerate(stranice):
            print(f"Ukupan broj stranica: {len(stranice)}")
            print("--------------------------------------------------")
            print(f"Obrađujem stranicu {broj_stranice + 1}...")

            # Koristimo srpski jezik za OCR (možete dodati i engleski ako želite)
            tekst = pytesseract.image_to_string(stranica, lang="eng+srp")

            izdvojeni_tekst += f"\n--- Stranica {broj_stranice + 1} ---\n{tekst}\n"

        # Čuvanje izdvojenog teksta u fajl
        with open(izlazni_txt_fajl, "w", encoding="utf-8") as f:
            f.write(izdvojeni_tekst)

        print(f"Tekst uspešno izdvojen i sačuvan u {izlazni_txt_fajl}")

    except Exception as e:
        print(f"Greška: {e}")


# Primer korišćenja:
pdf_fajl = "/home/datatab/Documents/NADSTRESNICA/nadstresnica/dokumentacija/06Projekat_decembar 2018.pdf"  # Promenite u putanju do vašeg PDF fajla
izdvoji_tekst_iz_skeniranog_pdf(pdf_fajl)
```
